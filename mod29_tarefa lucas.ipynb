## Nmero de grupos

Utilize as tcnicas vistas em aula que te ajudem a decidir pelo nmero de grupos a ser utilizados.

```python
variaveis = ["Administrative", "Administrative_Duration", "Informational", 
             "Informational_Duration", "ProductRelated", "ProductRelated_Duration"]
```

```python
df_pad = pd.DataFrame(StandardScaler().fit_transform(df[variaveis]), columns = df[variaveis].columns)
```

```python
df_pad.head()
```

```python
SQD = []
K = range(1,15)
for k in tqdm(K):
    km = KMeans(n_clusters=k)
    km = km.fit(df_pad)
    SQD.append(km.inertia_)
```

```python
SQD
```

```python
df_graf = pd.DataFrame({'num_clusters': list(range(1, len(SQD)+1)), 'SQD': SQD})
df_graf.plot(x = 'num_clusters', y = 'SQD');
```

```python
df_pad = pd.DataFrame(StandardScaler().fit_transform(df[variaveis]), columns = df[variaveis].columns)
```

```python
# Inicializar uma lista vazia que vai conter os valores da silueta mdia
siluetas = []
# Este  o nmero mximo de grupos que vamos testar
max_clusters = 10

# O loop percorre de 2 at o mximo de clusters
for n_clusters in tqdm(range(2, max_clusters+1)):
    # Aqui rodamos o k-means
    km = KMeans(n_clusters=n_clusters).fit(df_pad)
    # Aqui calculamos a silueta e guardamos o resultado na lista 'siluetas'
    siluetas.append(silhouette_score(df_pad, km.labels_))
```

```python
df_silueta = pd.DataFrame({'n_clusters': list(range(2, max_clusters+1)), 'silueta_mdia': siluetas})

df_silueta.plot.line(x = 'n_clusters', y = 'silueta_mdia', marker='o')
# plt.scatter(x = df_silueta.n_clusters, y = df_silueta.silueta_mdia)
```

## Avaliao dos grupos

Faa uma anlise descritiva para pelo menos duas solues de agrupamentos (duas quantidades diferentes de grupos) sugeridas no item anterior, utilizando as variveis que esto no escopo do agrupamento.
- Com base nesta anlise e nas anlises anteriores, decida pelo agrupamento final. 
- Se puder, sugira nomes para os grupos.

```python
df_pad = pd.DataFrame(StandardScaler().fit_transform(df[variaveis]), columns = df[variaveis].columns)

cluster_2 = KMeans(n_clusters=2)
cluster_2.fit_predict(df_pad)

df_pad['grupos_2'] = pd.Categorical(cluster_2.labels_)
```

```python
sns.pairplot(df_pad, hue='grupos_2')
```

```python
df_pad = pd.DataFrame(StandardScaler().fit_transform(df[variaveis]), columns = df[variaveis].columns)

cluster_3 = KMeans(n_clusters=3)
cluster_3.fit_predict(df_pad)

df_pad['grupos_3'] = pd.Categorical(cluster_3.labels_)
```

```python
sns.pairplot(df_pad, hue='grupos_3')
```

```python
df_pad = pd.DataFrame(StandardScaler().fit_transform(df[variaveis]), columns = df[variaveis].columns)

cluster_4 = KMeans(n_clusters=4)
cluster_4.fit_predict(df_pad)

df_pad['grupos_4'] = pd.Categorical(cluster_4.labels_)
```

```python
sns.pairplot(df_pad, hue='grupos_4')
```

Acredito que o modelo com 4 Clusters seja apropriado.

- O grupo 1 seria um grupo dos acessos iguais ou prximos de zero com pouco tempo de acesso tambm.

- O grupo 0 seria um grupo com poucos acessos no geral e pouco tempo de acesso em pginas informativas e de produtos, talvez um grupo administrativo.

- O grupo 3 seria um grupo de quantidade e tempo de acesso em pginas informativas.

- O grupo 2 seria um grupo de quantidade e tempo de acesso em pginas de produtos.
